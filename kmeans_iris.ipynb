```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn import datasets

%matplotlib inline
```

```python
df = pd.read_csv('iris.data', names=['sepal_l', 'sepal_w', 'length_l', 'length_w', 'target'])
# inicializar dados 
```

```python
df # exibir todos os dados
```

```python
# criar uma nova matriz sem a coluna de tipos de iris
X = np.array(df.drop('target', axis=1))
```

```python
kmeans = KMeans(n_clusters=3, random_state=0) # inicializar algoritmo
kmeans.fit(X) # treinar algoritmo
```

```python
df['k_classes'] = kmeans.labels_ # rotulos de agrupados
```

```python
# grafico de disperso original
origin = sns.pairplot(df, hue="target", vars=['sepal_l', 'sepal_w', 'length_l', 'length_w'])
```

```python
# exibir grafico em relacao ao agrupamento
origin = sns.pairplot(df, hue="k_classes", vars=['sepal_l', 'sepal_w', 'length_l', 'length_w'])
```
