```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.data', names=['sepal_l', 'sepal_w', 'length_l', 'length_w', 'target'])
```

```python
X = np.array(df.drop('target', axis=1))
```

```python
clustering = AgglomerativeClustering(n_clusters=3)
```

```python
clustering.fit(X)
```

```python
df['aglomerative'] = clustering.labels_
```

```python
origin = sns.pairplot(df, hue="target", vars=['sepal_l', 'sepal_w', 'length_l', 'length_w'])
```

```python
origin = sns.pairplot(df, hue="aglomerative", vars=['sepal_l', 'sepal_w', 'length_l', 'length_w'])
```
