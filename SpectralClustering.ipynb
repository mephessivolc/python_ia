```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('iris.data', names=['sepal_l', 'sepal_w', 'length_l', 'length_w', 'target'])
```

```python
X = np.array(df.drop('target', axis=1))
```

```python
feature = SpectralClustering(n_clusters=3, assign_labels="discretize",random_state=0)
```

```python
df['aglomerate'] = feature.fit(X).labels_
```

```python
origin = sns.pairplot(df, hue="target", vars=['sepal_l', 'sepal_w', 'length_l', 'length_w'])
```

```python
origin = sns.pairplot(df, hue="aglomerate", vars=['sepal_l', 'sepal_w', 'length_l', 'length_w'])
```
